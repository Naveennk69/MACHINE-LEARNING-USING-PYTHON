```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn import tree 
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics 
import seaborn as sn 
```

```python
data = pd.read_csv("C:/Users/USER/OneDrive/Desktop/G NAVEEN KUMAR GOWDA/DATA/DataSet/college_student_placement_dataset.csv")
data
```

```python
x=data.iloc[:,1:-1].values  
y=data.iloc[:,-1].values
print(x,y)
```

```python
label=LabelEncoder()
data['Internship_Experience']=label.fit_transform (data['Internship_Experience'])
data['Placement']=label.fit_transform (data['Placement'])
data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=5) 
print(x_train, x_test, y_train, y_test)
```

```python
model= DecisionTreeClassifier(criterion='entropy',random_state=5) 
model.fit(x_train, y_train) 
y_pred=model.predict(x_test) 
print('y_pred: ', y_pred) 
```

```python
conf_mat=metrics.confusion_matrix(y_test, y_pred)  
print('Confusion Matrix : ', conf_mat) 
Accuracy_score=metrics.accuracy_score(y_test, y_pred) 
print('Accuracy Score : ', Accuracy_score) 
print('Accuracy in Percentage : ', int(Accuracy_score*100),'%')
```

```python
conf_mat=pd.crosstab(y_test, y_pred, rownames=['Actual'], 
colnames=['Predicted']) 
sn.heatmap(conf_mat, annot=True)
```
