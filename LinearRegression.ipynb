```python
# Importing Required Libraries:
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
#Importing (Reading) Datasets:
data=pd.read_csv("C:/Users/HP/Desktop/New folder (3)/archive (4)/FINAL_USO.csv")
data
```

```python
label = LabelEncoder()

# Apply LabelEncoder to each column one by one
for col in ['Date', 'Open', 'High', 'Low', 'Close']:
    data[col] = label.fit_transform(data[col])

print(data)
```

```python
# Exploring Dataset:
data.head()
```

```python
data.shape
```

```python
#Checking for any null values in dataset:
data.isnull().sum()
```

```python
#Assigning Dependent and Independent variables:
x=data.iloc[:,:1].values
y=data.iloc[:, 1:2].values
print(x,y)
```

```python
# Splitting the dataset into Training and Testing Dataset:
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state = 42)
print(x_train, x_test, y_train, y_test)
```

```python
# Fitting the Model (Linear Regression):
model=LinearRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
print(y_pred)
```

```python
print(y_pred)
print(y_test)
```

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('Gold Price Prediction (training set)')
plt.xlabel('Low')
plt.ylabel('High')
plt.show()
```

```python
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_train, model.predict(x_train), color='red')
plt.title('Gold Price Prediction  (testing set)')
plt.xlabel('High')
plt.ylabel('Low')
plt.show()
```
