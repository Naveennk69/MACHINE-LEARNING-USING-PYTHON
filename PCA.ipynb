```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
```

```python
data = pd.read_csv("C:/Users/USER/OneDrive/Desktop/G NAVEEN KUMAR GOWDA/DATA/DataSet/college_student_placement_dataset.csv")
data
```

```python
data.head()
```

```python
data.shape
```

```python
label=LabelEncoder()
data['Internship_Experience']=label.fit_transform (data['Internship_Experience'])
data['Placement']=label.fit_transform (data['Placement'])
data.tail()
```

```python
x=data.iloc[:,1:-1].values  
y=data.iloc[:,-1].values
print(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=5)
print(x_train, x_test, y_train, y_test)
```

```python
from sklearn.decomposition import PCA 
principal=PCA(n_components=3) 
x=principal.fit_transform(x) 
print(x.shape) 
print(x)
```

```python
import matplotlib.pyplot as plt
principal.components_ 
plt.figure(figsize=(10,10)) 
plt.title('PCA 2D') 
plt.scatter(x[:,0],x[:,1],c=data['Placement'],cmap='plasma') 
plt.xlabel('pc1') 
plt.ylabel('pc2') 
```

```python
from mpl_toolkits.mplot3d import Axes3D 
fig=plt.figure(figsize=(10,10)) 
axis=fig.add_subplot(111,projection='3d') 
axis.scatter(x[:,0],x[:,1],x[:,2],c=data['Placement'],cmap='plasma') 
axis.set_xlabel('PC1',fontsize=10) 
```
